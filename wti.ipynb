```python
%matplotlib inline
import matplotlib.pyplot as plt
from datapackage import Package
import pandas as pd

package = Package('https://datahub.io/core/oil-prices/datapackage.json')

for resource in package.resources:
  if resource.name == "wti-daily_csv":
    df = pd.read_csv(resource.descriptor['path'],index_col='Date', parse_dates=True) 
    df = df.query("Date >= '2010-01-01'")
    ax = df.plot()
    plt.xticks(rotation=90)
    plt.legend(loc='best')
    plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from datapackage import Package
import pandas as pd

package = Package('https://datahub.io/core/bond-yields-us-10y/datapackage.json')

for resource in package.resources:
  if resource.name == "monthly_csv":
    df = pd.read_csv(resource.descriptor['path'], index_col=["Date"], parse_dates=True)
    df = df.query("Date >= '1980-01-01'")
    ax = df.plot()
    plt.xticks(rotation=90)
    plt.legend(loc='best')
    plt.show()
```
